# Two options for subplots

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
# past some data to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
# plot
ax[0, 0].plot(x, x**4);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# OO Method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
over_50.shape
```

```python
# pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10, 6));
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax); 
ax.set_xlim([50, 80]); # adjust the x axis limit
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);

# customise the plot
ax.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholesterol');

# add a legend
ax.legend(*scatter.legend_elements(), title='target');

# add mean horizontal line
ax.axhline(over_50['chol'].mean(), # set the mean line
          linestyle='--'); # use dash line
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10));

# add data to ax
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])

# customise ax
ax0.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')
ax1.set(title='Heart Disease and Max Heart Rate Levels',
       xlabel='Age',
       ylabel='Max Heart Rate')

# add a legend
ax0.legend(*scatter.legend_elements(), title='Target')
ax1.legend(*scatter.legend_elements(), title='Target')

# add a mean line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--');
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--');
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True); ###

# add data to ax
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])

# customise ax
ax0.set(title='Heart Disease and Cholesterol Levels',
       ylabel='Cholesterol')
ax1.set(title='Heart Disease and Max Heart Rate Levels',
       xlabel='Age',
       ylabel='Max Heart Rate')

# add a legend
ax0.legend(*scatter.legend_elements(), title='Target')
ax1.legend(*scatter.legend_elements(), title='Target')

# add a mean line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--');
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

# add a title to the figure
fig.suptitle('Heart Disease Analysis', 
             fontsize=16, 
             fontweight='bold');
```

### customising x and y axis limit

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True); ###

# add data to ax
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='winter')

# customise ax
ax0.set(title='Heart Disease and Cholesterol Levels',
        ylabel='Cholesterol')
ax1.set(title='Heart Disease and Max Heart Rate Levels',
       xlabel='Age',
       ylabel='Max Heart Rate')

# change the x axis limit
ax0.set_xlim([50, 80]) 
# ax1 doesn't need a change since both share the same x axis

# change the y axis limit
ax1.set_ylim([60, 200])

# add a legend
ax0.legend(*scatter.legend_elements(), title='Target')
ax1.legend(*scatter.legend_elements(), title='Target')

# add a mean line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--');
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

# add a title to the figure
fig.suptitle('Heart Disease Analysis', 
             fontsize=16, 
             fontweight='bold');
```
