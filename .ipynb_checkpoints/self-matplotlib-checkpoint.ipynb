```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot()
```

```python
plt.plot([1,2,3,4]);
```

```python
x = [1, 2,3 , 4]
y=[11, 22, 33, 44]
plt.plot(x, y);
```

### 1st method

```python
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

### 2nd method

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)# add some data
plt.show()
```

### 3rd Method (recommended)

```python
fig, ax = plt.subplots()
ax.plot(x, y);
```

### Maplotlib example workflow

```python
# import matplotlib

# Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plot data
ax.plot(x, y)

# Customise plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# save and show
fig.savefig('./images/sample-plot.png')
```

### Making figures with NumPy arrays

```python
# create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# use the same data to create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), 
       height=nut_butter_prices.values());
ax.set(title="Nut butter price",
      ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), 
        list(nut_butter_prices.values()));
```

```python
# make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
# past some data to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
# plot
ax[0, 0].plot(x, x**4);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### OR

```python
car_sales = pd.read_csv('./pandas-materials/car-sales.csv')
car_sales.head()
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '')
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str[:-2]
car_sales
```

```python
car_sales['Sale Date'] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].astype('int64')
type(car_sales['Price'][0])
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales
```

```python
# Plot the total sales overtime
car_sales.plot(x="Sale Date", y="Total Sales");
```
